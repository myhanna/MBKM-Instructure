```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.family'] = 'monospace'
plt.rcParams['font.size'] = 15

fig = plt.figure(figsize=[12,8])

axBnd = fig.add_axes([0.00, 0.0, 0.30, 0.4])
axDOS = fig.add_axes([0.35, 0.0, 0.15, 0.4])

prefix     = 'si'
folderBnd  = 'bands/'
folderDOS  = 'dos/'

EFermi =  6.255

#--------------------------------------------------------------#
# Bands Structure
#--------------------------------------------------------------#

nBand = 8
nkpts = 91 # liat di bands.gnu nya liat k point nya 

fBand   = open(folderBnd+prefix+'.bands.gnu', 'r')
dataBnd = [line for line in fBand.readlines() if line.strip()]
fBand.close()

kpts = []
for ik in range (nkpts):
    kpts.append(float(dataBnd[ik].split()[0]))
    
print(len(kpts))
      
band = []
for ib in range (nBand):
    band.append([])
    for ik in range (nkpts):
        band[ib].append(float(dataBnd[ib*nkpts+ik].split()[1]))

print(len(band))

# Data band  untuk tiap kpoint 
eBnd = []
for ik in range (nkpts):
    eBnd.append([])
    for ib in range (nBand):
        eBnd[ik].append(band[ib][ik]-EFermi)
#k-path
L  = kpts[0]
r1  = kpts[20]
X  = kpts[50]
U  = kpts[60]
r2  = kpts[90] 

axBnd.set_xlim(r1,r2)
axBnd.set_ylim(-7,7)

axBnd.set_xticks([L,r1,X,U,r2])
axBnd.set_xticklabels(['L','$\Gamma$','X','U','$\Gamma$'])

axBnd.set_ylabel ('Energy (eV)')
axBnd.yaxis.set_minor_locator(mpl.ticker.MultipleLocator(2.0))
axBnd.yaxis.set_minor_locator(mpl.ticker.MultipleLocator(0.4))

axBnd.yaxis.set_tick_params(which='major',size =6, direction = 'in', right='on')
axBnd.yaxis.set_tick_params(which='minor',size =4, direction = 'in', right='on')

axBnd.xaxis.set_tick_params(which='major', bottom=False, top=False)

axBnd.axvline(L, color ='black', linewidth = 1.5)
axBnd.axvline(r1, color ='black', linewidth = 1.5)
axBnd.axvline(X, color='black', linewidth  = 1.5)
axBnd.axvline(U, color='black', linewidth  = 1.5)
axBnd.axvline(r2, color ='black', linewidth = 1.5)


axBnd.set_title('(a)',loc = 'left', pad = 10)

axBnd.plot(kpts,eBnd, color='darkblue')

        
#--------------------------------------------------------------#
# DOS
#--------------------------------------------------------------#

eDOS, DOS = np.loadtxt(folderDOS+prefix+'.dos', usecols=(0,1), unpack=True)
maxDOS = np.max(DOS)

axDOS.set_xlim(0,maxDOS)
axDOS.set_ylim(-7,7)

axDOS.set_xticks([])
axDOS.set_xlabel('DOS (a.u.)')
axDOS.axhline(0,color ='black', linewidth=1.0, linestyle = '--')

axDOS.yaxis.set_minor_locator(mpl.ticker.MultipleLocator(2.0))
axDOS.yaxis.set_minor_locator(mpl.ticker.MultipleLocator(0.4))

axDOS.yaxis.set_tick_params(which='major',size =6, direction = 'in', right='on')
axDOS.yaxis.set_tick_params(which='minor',size =4, direction = 'in', right='on')

axDOS.set_title('(b)', loc = 'left', pad = 10)

axDOS.xaxis.set_tick_params(which='major', bottom=False, top=False)
axDOS.plot(DOS,eDOS-EFermi, color = 'darkblue')

axDOS.fill_betweenx(eDOS-EFermi, 0, DOS, facecolor ='darkblue', alpha = 0.2)

#plt.show()

plt.savefig('DOS,BANDS silikon.pdf', bbox_inches='tight')
```

```python
import matplotlib.pyplot as plt
plt.style.use('style/sci.mplstyle')
import numpy as np

efermi = 6.255
def r_dos(name):
    ener, dos = np.loadtxt(name, usecols=(0,1),
        unpack = True)
    return ener, dos

# read the total pdos
ener, dos = r_dos('DOS/si.pdos_tot')
# read the Pdos of C atom number 1
ener1, p1Cs = r_dos('DOS/si.pdos_atm#1(Si)_wfc#1(s)')
ener1, p1Cp = r_dos('DOS/si.pdos_atm#1(Si)_wfc#2(p)')
ener1, p2Cs = r_dos('DOS/si.pdos_atm#2(Si)_wfc#1(s)')
ener1, p2Cp = r_dos('DOS/si.pdos_atm#2(Si)_wfc#2(p)')

#Create figure object
plt.figure()
#Plot the dos
plt.plot(dos,efermi-ener, c = 'k')
#plot the PDOS of s- orbital
plt.plot(p1Cs+p2Cs, efermi-ener, c = 'b')
#plot the PDOS of p- orbital
plt.plot(p1Cp+p2Cp, efermi-ener, c = 'r')
#Add the x and y-axis labels
plt.ylabel('Energy (eV') # di ubah
plt.xlabel('PDOS (state/eV/unit-cell)') # di ubah 
# set the axis limits
plt.xlim(0,3)
plt.ylim(-10,10)
#Save the Figure
plt.savefig('plot-pdos.pdf')
# show the figure
plt.show()
```
