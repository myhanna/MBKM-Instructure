```python
import matplotlib.pyplot as plt
import numpy as np
from qutip import (coherent, destroy, expect, mesolve, HarmonicOscillatorWaveFunction)
```

```python
N = 35
w = 2 * np.pi
tlist = np.linspace(0, 10, 101)
```

```python
a = destroy(N)
```

```python
H = w * a.dag() * a
```

```python
psi0 = coherent(N, 2.0)
```

```python
result = mesolve(H, psi0, tlist)
```

```python
#AYOK GIMANA YOROBUN

fig = plt.figure()

def V(x):
    return 0.5 * x ** 2

xmin = -3
xmax = 3
xlist = np.linspace(xmin, xmax, 100)

for i in range(0, len(tlist)):

    ax = plt.axes(xlim = (xmin, xmax),  ylim = (-2, 2))

    fig.suptitle(f'$t = {round(tlist[i], 2)}$')

    ax.plot(xlist, V(xlist), label= r"$v(x)$")
    ax.legend(loc = "best")

    ax.axhline(y = 0, ls = ":", lw = 0.5, c = 'k')

    psi_x = HarmonicOscillatorWaveFunction(psi = result.states[i], omega = w, extent = [xmin, xmax])

    psi_x.data = np.abs(psi_x.data)

    psi_x.visualize(fig = fig, ax = ax, show_ylabel= False)

    plt.savefig(f"pics/{i:03d}.png", dpi = 600)
```
