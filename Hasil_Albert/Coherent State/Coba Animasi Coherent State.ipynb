```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# Set common figure parameters:
newparams = {'axes.labelsize': 14, 'axes.linewidth': 1, 'savefig.dpi': 300,
             'lines.linewidth': 1.0, 'figure.figsize': (8, 3),
             'figure.subplot.wspace': 0.4,
             'ytick.labelsize': 10, 'xtick.labelsize': 10,
             'ytick.major.pad': 5, 'xtick.major.pad': 5,
             'legend.fontsize': 10, 'legend.frameon': False,
             'legend.handlelength': 1.5}
plt.rcParams.update(newparams)
```

```python
n = 1000       # number of points per unit on the x-axis.
p = 10          # the number of points included on the x-axis

x = np.linspace(-p,p,p*n,True)
V = 0.5*x**2
omega = 6
alpha = 2
hbar = 1
xi = x*np.sqrt(omega/hbar)
n = 0
for t in np.arange(0, 10, 0.10):
    psi = ((omega/(np.pi*hbar))**(0.25)*
          np.exp(-np.abs(alpha)**2/2)*
          np.exp(xi**2/2)*
          np.abs(np.exp(-(xi-alpha*np.exp(-1j*omega*t)/np.sqrt(2))**2))
          )
    #Plot:
    fig, ax = plt.subplots(figsize=(7,4))
    analitik = ax.plot(x, psi, color='blue', label = 'Analytical wavefunction') # analytical eigenfunction
    potensial = ax.plot(x, V, color='red', label = r'Potential $V(x)$')         # potential
    ax.legend(['Analytical wavefunction', r'Potential $V(x)$'], loc = 'upper right', fontsize="8")
    ax.set_xlabel(r'$x$')
    ax.set_ylabel(r'$\psi(x)$')
    ax.set_ylim([-0.5, 3])
    ax.set_xlim([-8, 9])
    fig.suptitle(f't = {t}', fontsize=16)
    fig.savefig(f'./Gambar/gambar:{n:02d}.png', dpi=600)
    n += 1
    plt.show()
```
