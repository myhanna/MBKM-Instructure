```python
import numpy as np
import matplotlib.pyplot as plt
import qutip as qt
options = qt.Options(nsteps = 1e6)
```

```python
N = 35
w = 2 * np.pi
tlist = np.linspace(0, 2, 3)

a = qt.destroy(N)

Ham = w * a.dag() * a
```

Coherent states

```python
psi_0 = qt.coherent(N, 2.0)

psi_t = qt.mesolve(Ham, psi_0, tlist, options = options).states

fig, ax = plt.subplots(1, 2, figsize = (10, 5))

for psi in psi_t:
    ax[0].cla()
    qt.plot_wigner_fock_distribution(psi, fig = fig, axes = ax)
```

Squeezed vacuum

```python
psi_0 = qt.squeeze(N, 1.0) * qt.basis(N, 0)

psi_t = qt.mesolve(Ham, psi_0, tlist, options = options).states

fig, ax = plt.subplots(1, 2, figsize = (10, 5))

for psi in psi_t:
    ax[0].cla()
    qt.plot_wigner_fock_distribution(psi, fig = fig, axes = ax)
```

Squeezed coherent state

```python
psi_0 = qt.displace(N, 2.0) * qt.squeeze(N, 1.0) * qt.basis(N, 0)

psi_t = qt.mesolve(Ham, psi_0, tlist, options = options).states

fig, ax = plt.subplots(1, 2, figsize = (10, 5))

for psi in psi_t:
    ax[0].cla()
    qt.plot_wigner_fock_distribution(psi, fig = fig, axes = ax)
    plt.show()
```
