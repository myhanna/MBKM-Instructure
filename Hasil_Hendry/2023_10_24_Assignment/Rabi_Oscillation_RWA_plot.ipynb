```python
import numpy as np
import matplotlib.pyplot as plt
```

# **_Plotting Rabi transition_**

The equation to plot:
$$
P_{i\rightarrow f} = \frac{\left|V_{fi}\right|^2}{\hbar^2}\left( \frac{\sin^2\left(\frac{\Delta t}{2}\right)}{\Delta^2} \right)
$$

Let
$$
A = \frac{\left|V_{fi}\right|^2}{\hbar^2}
$$

The period for one compelete oscillation is given by
$$
T = \frac{2\pi}{(\Delta/2)}
$$

Figure 4.2 plots as a function of $\Delta$ but $\Delta$ is specified to be inversely proportional to $t$.
$$
\Delta\propto\frac{1}{t}
$$


```python
Delta_param = 0.05
A = 1

num_of_wavelength = 0.8
period = 2 * np.pi / (Delta_param / 2)

t_list = np.linspace(0, num_of_wavelength * period, 1000)

def P_function_of_t(t, Delta_param):
    return A * (np.sin(Delta_param * t / 2) / Delta_param) ** 2
    
#################################################################

Delta_list = np.linspace(-6 * np.pi, 6 * np.pi, 1000)

def P_function_of_Delta(Delta):
    return A * (np.sin(Delta / 2) / Delta) ** 2

#####################################################################

fig, ax = plt.subplots(1, 2, figsize = (9, 4))
fig.suptitle("Rabi Oscillation under the Rotating Wave Approximation")

p = ax[0]
p.plot(t_list, func := P_function_of_t(t_list, Delta_param))
p.set_xlim(t_list[0], t_list[-1])
p.set_ylim(0, np.max(func) * 1.2)

num_of_wavelength = 0.18
Delta_param = 1e-7
t_list = np.linspace(0, num_of_wavelength * period, 1000)
p.plot(t_list, P_function_of_t(t_list, Delta_param), ls = "--")

p.annotate(text = r"$P^{(1)}_{i\rightarrow f}$", xy = (-0.12, 0.9), xycoords = "axes fraction", fontsize = 15)
p.annotate(text = r"$\Delta\approx 0$", xy = (0.035, 0.64), xycoords = "axes fraction", fontsize = 15)
p.annotate(text = r"$\Delta\mathrm{\ large}$", xy = (0.38, 0.8), xycoords = "axes fraction", fontsize = 15)
p.annotate(text = r"$t$", xy = (0.48, -0.07), xycoords = "axes fraction", fontsize = 15)

p = ax[1]
p.plot(Delta_list, func := P_function_of_Delta(Delta_list))
p.set_xlim(Delta_list[0], Delta_list[-1])
p.set_ylim(0, np.max(func) * 1.1)
p.axvline(x = 0 , c = "k", lw = 0.8)
p.annotate(text = r"$\frac{4\left|\left(\hat{\mathbf{d}}\cdot\mathbf{E}\right)_{fi}\right|^2}{\hbar^2}$", xy = (0.27, 0.9), xycoords = "axes fraction", fontsize = 15)
p.annotate(text = r"$\Delta$", xy = (1.02, -0.02), xycoords = "axes fraction", fontsize = 15)

for i in range(7):
    text = r"$%s\pi/t$" %(-6+i*2)
    if i < 3:
        xfrac = (i-0.1) * 1/7
    elif i > 3:
        xfrac = (i + 0.4) * 1/7
    else:
        xfrac = (i + 0.2) * 1/7
        
    p.annotate(text = text, xy = (xfrac, -0.05), xycoords = "axes fraction", fontsize = 12)

for i in range(2):
    ax[i].set_xticks(ticks = [])
    ax[i].set_yticks(ticks = [])

plt.tight_layout()
plt.savefig("Rabi_Oscillation_RWA_plot.jpg", dpi = 600)
plt.show()
```
